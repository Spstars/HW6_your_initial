```python
import csv
import matplotlib.pyplot as plt
f= open('seoul.csv','r',encoding='ANSI')
data=csv.reader(f)
header=next(data)
result=[]
for row in data:
    if row:
        if row[-1]!='':
            if row[0].split("-")[1]=="08":
                result.append(float(row[-1]))
f.close()
plt.plot(result,"hotpink")
plt.show()
```

```python
f= open('seoul.csv','r',encoding='ANSI')
data=csv.reader(f)
header=next(data)
avg=[]
for row in data:
    if row:
        month=row[0].split("-")[1]
        # print(month)
        if row[-1] !="":
            if month =='08':
                avg.append(float(row[-1]))
print(avg)
plt.hist(avg,color="r")
plt.show()
```

```python
f=open("seoul.csv",'r',encoding='ANSI')
data=csv.reader(f)
next(data)

month=[[],[],[],[],[],[],[],[],[],[],[],[],]
print(month)
for row in data:
    if row and row[-1] !="":
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
plt.boxplot(month)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f=open('gender.csv','r',encoding="ANSI")
data=csv.reader(f)

header=next(data)
# print(header)
male=[]
female=[]

for row in data:
    if '신도림' in row[0]:
        print(row[0])
        for i in range(0,101):
            male.append(int(row[i+3]))
            female.append(int(row[-(i+1)]))
print(female)
female.reverse()
print(female)
print(male)
f.close()
```

```python
f10=female[:10]
m10=male[:10]
m10=list(map(lambda x: -x,m10))
print(m10)
plt.barh(range(10),m10)
plt.barh(range(10),f10)
```
